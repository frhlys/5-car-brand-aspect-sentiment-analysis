```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from requests import get
from urllib.request import urlopen
from bs4 import NavigableString
import re
import time
from lxml.html import fromstring
import requests
from itertools import cycle
import traceback

car_model={'A3'
,'A4'
,'A5'
,'A6'
,'A7'
,'Q5 Sportback'
,'Q4 Sportback e-tron'}

car_type={'suv', 'sedan', 'hatchback', 'wagon'}
#'coupe','wagon','sedan','hatchback','suv','type-s','32-type-s','hybrid','type-r'}
car_year={'2019','2020','2021','2022','2023','2024' }

Review_Title=[]
Vehicle_Rating=[]
Vehicle_Name=[]
Review=[]
AuthorName=[]
Review_Date=[]

for model in car_model:
    for year in car_year:
        for c_type in car_type:
            for pages in range(1,15):
                
                try:
                    time.sleep(1)
                    print(model)
                    print(year)
                    print(c_type)
                    print(pages)
                    url = 'https://www.edmunds.com/audi/'+ model + '/'+ year + '/' + c_type + '/' + 'consumer-reviews/pg-' + str(pages)
                    response = requests.get(url)
                    if response.status_code != 200:
                        break
                    html_soup = BeautifulSoup(response.text, 'lxml')

                    review_widget = html_soup.find('div',attrs={'class':'body-reviews-widget'})
                    individual_review_container = review_widget.find_all(class_='individual-review-container')
                    
                    for a in individual_review_container: 
    
                        individual_overall_rating= a.find_all('div',{'class':'individual-overal-rating'})
    
                        for b in individual_overall_rating:
        
                            review_title=b.find('span',attrs={'class':'reviewTitle'}).text
                            consumer_rating_overall=b.find('div',{'class':'consumer-reviews last new-rating-star'}).find('span',{'class':'rating-big'})['title']
        
                            Review_Title.append(review_title)
                            Vehicle_Rating.append(consumer_rating_overall)
        
                            vehicle_title = a.find('div',attrs={'style':'float:left'}).find('p').text
                            Vehicle_Name.append(vehicle_title)
    
                            review=a.find_all('p',{'class':'review-text'})[1].text
                            Review.append(review)
    
                            author_heading = a.find_all(class_='author-heading')
    
                            for c in author_heading:
        
                                author_name=c.find(class_='author-crr').text
                                author_date=c.find_all(class_='cr-author')[1].text
        
                                AuthorName.append(author_name)
                                Review_Date.append(author_date)
                                
                except Exception as e:
                    # If the download for some reason fails (ex. 404) the script will continue downloading
                    # the next article.
                    print(e)
                    print("continuing...")
                    continue


df = pd.DataFrame(
{'Review_Date': Review_Date,
    'Author_Name': AuthorName,
    'Vehicle_Title': Vehicle_Name,
    'Review_Title': Review_Title,
    'Review': Review,
    'Rating': Vehicle_Rating
})
    
df.to_csv('Scrapped_Car_Review_audi.csv')
```
